# Handle Categorical Data

```python
# Machine learning models cannot understand text
# Here it is Location
# convert categories → numbers
# location col must be encoded
```

```python
# One-Hot Encoding
# sklearn - scikit-learn => "Scientific Kit for Learning"
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
# Create encoder object
# OneHotEncoder()	create encoder object
# encoder	variable storing that object
encoder = OneHotEncoder()
```

```python
# encoded_location = encoder.fit_transform((df["Location"])) === error
 
# df["Location"] → Series (1D) ====== NO
# OneHotEncoder needs → DataFrame / 2D array ==== YES
```

```python
# double brackets so pandas sends a DataFrame (2D)

# Fit encoder and transform Location column

#| fit()       | learn the categories           |
#| transform() | convert them to binary columns |

encoded_location = encoder.fit_transform(df[["Location"]])
```

```python
#Categories found :
# Rural
# Suburban
# Urban
# O/P becomes
# [[0 0 1]
# [0 0 1]
# [0 1 0]]

# Sparse Matrix stores only non-zero values to save memory
# (0, 2)  1.0 [Row index = 0 | Column index = 2  | Value = 1] -- [0 0 1]

print(encoded_location)
print("---------------")
print(encoded_location.toarray())
```

```python
# Convert encoded result to DataFrame with proper column names
# encoded_df = pd.DataFrame(encoded_location.toarray())

encoded_df = pd.DataFrame(
    encoded_location.toarray(),
    columns = encoder.get_feature_names_out(["Location"])
)
```

```python
# pd.DataFrame() converts data into a table format (rows & columns)
# encoded data into a proper table

# encoded_location.toarray() === convert sparse matrix → normal array
# OneHotEncoder returned a Sparse Matrix
# sparse format -- (0,2) 1
# but DataFrames need normal array data
# we convert it to a NumPy array

# columns
# we don’t give it, pandas will create 0 1 2
# get_feature_names_out : generates column names automatically based on original column
```

```python
# Remove Original Location Column
df = df.drop("Location", axis=1)
```

```python
df
```

```python
# Combine Encoded Columns
df = pd.concat([df,encoded_df],axis=1)
```

```python
df
```

```python
X = df[["Size_sqft","Bedrooms","Age","Location_Rural","Location_Suburban","Location_Urban"]]
X.head()
```

```python
y = df["Price"]
y.head()
```

# How OneHotEncoder works

```python
# OneHotEncoder reads the data, it automatically sorts them alphabetically
# if col has [ Urban Rural Suburban ]
# then order becomes [ Rural Suburban Urban ]

# For each row
# Put 1 in the column that matches the category
# Put 0 in all other columns
#| Location | Rural | Suburban | Urban |
#| -------- | ----- | -------- | ----- |
#| Urban    | 0     | 0        | 1     |
#| Rural    | 1     | 0        | 0     |
#| Suburban | 0     | 1        | 0     |

# Rural     → [1 0 0]
# Suburban  → [0 1 0]
# Urban     → [0 0 1]
```

# Feature Scaling

```python
# Our dataset has different range values
# Large values like 2000 dominate smaller values like 3 or 1

# Size_sqft	2000
# Bedrooms	3

# We use standardization
# Z = (X − mean) / standard deviation
# After scaling, values look like: -1.3, 0.4, 1.2 (similar scale)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X.head()
```

```python
# fit()	learn mean & std from data
# transform()	scale the data
# fit_transform()	do both together
X_scaled = scaler.fit_transform(X)
```

```python
print(X_scaled)
```

```python
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
print(X_scaled_df)
```

```python
# We usually do NOT scale one-hot encoded columns
# for learning purposes it's fine if you scaled them
```

# Train–Test Split

```python
# Training Data	Used to train the model - 80%
# Testing Data	Used to test the model - 20%
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Random_state=42  == fixed random split
# shuffled randomly each time
# RUN 1 :: Train: rows 1,3,4,6,7... | Test : rows 2,5 (random 80% test and random 20% test)
# RUN 2 :: Train: rows 2,4,5,7... | Test : rows 1,6 

# random_state = 42 means each run
# Train: same rows
# Test : same rows
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size = 0.2, random_state = 42
)
```

```python
# Dataset = 15 rows (20% of 15 = 3 for Testing | 12 for Training)

print(X_train.shape) # 12 rows 6 cols
print(X_test.shape) # 3 rows 6 cols
print(y_train.shape) # 12 target values
print(y_test.shape) # 3 target values
```

# Train the Machine Learning Model

```python
# Only Regression models : predict continuous values

# Classification : used when the target is categories.
# Ex: Spam Detection ->	Spam / Not Spam
```

# Linear Regression

```python
#x1 = Size_sqft
#x2 = Bedrooms
#x3 = Age
#x4 = Location_Rural
#x5 = Location_Suburban
#x6 = Location_Urban

#Price =
#b0 +
#b1*(Size_sqft) +
#b2*(Bedrooms) +
#b3*(Age) +
#b4*(Location_Rural) +
#b5*(Location_Suburban) +
#b6*(Location_Urban)

#Size_sqft = 1700
#Bedrooms = 3
#Age = 7
#Location_Urban = 1
#Location_Suburban = 0
#Location_Rural = 0

#Predicted Price =
#b0 +
#b1*1700 +
#b2*3 +
#b3*7 +
#b4*0 +
#b5*0 +
#b6*1

# o/p values
#277539.63
#267622.48
#176614.27
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
# Train Model
model.fit(X_train, y_train)
```

```python
model.coef_
model.intercept_
print(model.coef_)
print(model.intercept_)
```

```python
y_pred = model.predict(X_test)
```

```python
print(y_test.values)
print(y_pred)
```

```python
y_pred = y_pred.round().astype(int)
print(y_pred)
```

```python
# Evaluate Model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_test)
```

```python
# Measures average prediction error
# Lower = better.
```

```python
print("Mean squared error: ",mse)
print("R2 Score", r2)
```

```python
#| R²   | Meaning       |
#| ---- | ------------- |
#| 1.0  | Perfect model |
#| 0.8+ | Very good     |
#| 0.5  | Moderate      |
#| 0    | Poor          |
```

# Decision Tree

```python
# Decision Tree splits the data into rules
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
# Create Model
dt_model = DecisionTreeRegressor(random_state= 42)
```

```python
# Train model
dt_model.fit(X_train, y_train)
```

```python
# Make Predictions
y_pred_dt = dt_model.predict(X_test)
```

```python
print(y_test.values)
print(y_pred_dt)
```

```python
# Evaluate model
mse_dt = mean_squared_error(y_test, y_pred_dt)
r2_dt = r2_score(y_test,y_pred_dt)
```

```python
print("Mean squared error: ",mse_dt)
print("R2 Score", r2_dt)
```

```python
# Decision Trees usually need more data to learn good splitting rules
# so Predictions become unstable

# Linear regression r2=1 | better model
# Bigger house → Higher price
```

# Random Forest

```python
# Random Forest Regressor
# it fixes the biggest problem of Decision Trees: Overfitting
# Random Forest = many decision trees combined.
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
# Create Model
# n_estimators=100 :: number of trees in forest
rf_model = RandomForestRegressor(
    n_estimators = 100,
    random_state=42
)
```

```python
# Train the Model
rf_model.fit(X_train,y_train)
```

```python
# Make Predictions
y_pred_rf = rf_model.predict(X_test)
```

```python
print(y_test.values)
print(y_pred_rf)
```

```python
mse_rf = mean_squared_error(y_test,y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)
```

```python
print("Mean squared error: ",mse_rf)
print("R2 Score", r2_rf)
```

```python
#Random Forest works using a method called Bagging (Bootstrap Aggregation):
#Multiple Trees
#     ↓
#Different subsets of data
#     ↓
#Average predictions
#     ↓
#Better accuracy
```
